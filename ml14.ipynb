```python


# Create a box plot
plt.boxplot(df['total ppl'])

Q1 = df['total ppl'].quantile(0.25)
Q3 = df['total ppl'].quantile(0.75)
IQR = Q3 - Q1

# Filter the DataFrame to remove outliers
df = df[(df['total ppl'] >= Q1 - 1.5 * IQR) & (df['total ppl'] <= Q3 + 1.5 * IQR)]
```

```python
plt.boxplot(df['illiterate'])

Q1 = df['illiterate'].quantile(0.25)
Q3 = df['illiterate'].quantile(0.75)
IQR = Q3 - Q1

# Filter the DataFrame to remove outliers
df = df[(df['illiterate'] >= Q1 - 1.5 * IQR) & (df['illiterate'] <= Q3 + 1.5 * IQR)]
```

```python
plt.boxplot(df['graduate n above'])

Q1 = df['graduate n above'].quantile(0.25)
Q3 = df['graduate n above'].quantile(0.75)
IQR = Q3 - Q1

# Filter the DataFrame to remove outliers
df = df[(df['graduate n above'] >= Q1 - 1.5 * IQR) & (df['graduate n above'] <= Q3 + 1.5 * IQR)]
```

```python
len(df)
```

```python
df['%illiterate']=df['illiterate']/df['total ppl']*100
```

```python
df['%graduate']=df['graduate n above']/df['total ppl']*100
```

```python
df
```

```python
df.drop('Area Name',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
standard_scaler=StandardScaler()
df=standard_scaler.fit_transform(df)
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
mergings = linkage(df, method = "single", metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
mergings = linkage(df, method = "complete", metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
df= pd.DataFrame(df)
df.columns = ['total ppl', 'illiterate', 'graduate n above', '%illiterate', '%graduate']
df.head()
```

```python
clusterCut = pd.Series(cut_tree(mergings, n_clusters=5).reshape(-1))
df_hc = pd.concat([df, clusterCut], axis=1)
df_hc.columns = ['total ppl', 'illiterate', 'graduate n above', '%illiterate', '%graduate','ClusterID']
```

```python
km_clusters_total_ppl = df_hc.groupby(["ClusterID"])["total ppl"].mean()

km_clusters_illiterate = df_hc.groupby(["ClusterID"])["illiterate"].mean()

km_clusters_graduate_n_above = df_hc.groupby(["ClusterID"])["graduate n above"].mean()

km_clusters_percent_illiterate = df_hc.groupby(["ClusterID"])["%illiterate"].mean()

km_clusters_percent_graduate = df_hc.groupby(["ClusterID"])["%graduate"].mean()

```

```python
df = pd.concat([pd.Series([0,1,2,3,4]),km_clusters_total_ppl , km_clusters_illiterate,km_clusters_graduate_n_above,km_clusters_percent_illiterate,km_clusters_percent_graduate], axis=1)
df.columns = ["ClusterID", "total ppl_mean", "illiterate_mean", "graduate_n_above_mean",'percent_illiterate_mean','percent_graduate_mean']
df.head()
```
